```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np
```

```python
features, true_labels = make_blobs(
    n_samples=200,
    centers=4,
    cluster_std=1.05,
       random_state=42
    )
```

```python
customPalette = ['#630C3A', '#39C8C6', '#D3500C', '#FFB139']

plt.scatter(features[:,0], features[:,1], c=true_labels)
labels = {'AAA': (-10.0,5.0),
          'AA': (-2.5,5.0),
          'A': (5.0,5.0),
          'BBB': (-7.5, -3.5)}
plt.title("Clusters Under Principle Components (Investment Grade)")
plt.xlabel("PC1")
plt.ylabel("PC2")
for i, label in enumerate(labels.keys()):
    plt.annotate(label, 
                 labels[label],
                 horizontalalignment='center',
                 verticalalignment='center',
                 size=20, weight='bold',
                 color=customPalette[i]) 
# features
```
