## KNN with CCA Features

```python
def plot_knn(train_x, train_y, test_x, test_y, k=30):
    trains = []
    tests = []
    trains1 = []
    tests1 = []
    upper_k = k

    fig = plt.figure(1, figsize=(10,10))
    ax = fig.add_subplot(1,1,1)
    for k in range(2, upper_k):
        for w in ['uniform', 'distance']:
            neigh = KNeighborsClassifier(n_neighbors=k, weights=w)
            neigh.fit(train_x, train_y)
    #         print(k, 'train: ',neigh.score(train_x, train_y),'test: ', neigh.score(test_x, test_y))
            if w == 'uniform':
                trains.append(neigh.score(train_x, train_y))
                tests.append(neigh.score(test_x, test_y))
            else:
                trains1.append(neigh.score(train_x, train_y))
                tests1.append(neigh.score(test_x, test_y))

    ax.plot(np.arange(2,upper_k),trains, label='train accuracy (uniform)')
    ax.plot(np.arange(2,upper_k),tests,  label='test accuracy (uniform)')

    ax.plot(np.arange(2,upper_k),trains1, label='train accuracy (distance)')
    ax.plot(np.arange(2,upper_k),tests1,  label='test accuracy (distance)')
    ax.set_xlim((2, upper_k))
    ax.legend()
    # ax.set_xlim((2,15))
    # len(np.arange(2,15))
    # ,tests
    ax.set_ylabel('Accuracy')
    ax.set_xlabel('K')
    plt.show()
    
```

```python
def score_one_off(clf, preds, true):
    """ 
    preds/true: must be array of int
    """
    assert len(preds) == len(true)
    bools = np.zeros(len(preds))
    for i in range(len(preds)):
        bools[i] = (preds[i] == true[i]) | (preds[i] == (true[i] + 1)) | (preds[i] == (true[i] - 1)) 
    return np.mean(bools)
    
    
    
    
```
